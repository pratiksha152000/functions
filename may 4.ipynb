## step4: Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## 4a: Univariate analysis for categorical features

```python
df['species'].value_counts()
```

```python
df['species'].value_counts().plot(kind='bar',title='species count plot')
```

```python
for i in cat:
    df[i].value_counts().plot(kind='bar',title=f'countplot for{i}')
    plt.show()
```

## 4b. Univariate analysis for continuous features

```python
con
```

```python
sns.histplot(data=df,x='sepal_length',kde=True)
plt.title('Histogram for sepal_length')
```

```python
con
```

```python
for i in con:
    sns.histplot(data=df,x=i,kde=True)
    plt.title(f'Histogram for {i}')
    plt.show()
    print('\n')
```

```python
x = 1
plt.figure(figsize=(10,10))
for i in con:
    plt.subplot(2,2,x)
    x = x+1
    sns.histplot(data=df,x=i,kde=True)
    plt.title(f'Histogram for {i}')

# Show subplot
plt.show()

```
